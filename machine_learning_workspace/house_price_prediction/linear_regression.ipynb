```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from holoviews.plotting.bokeh.styles import font_size, color
from sklearn import linear_model
```

```python
df = pd.read_csv('Home Prices.csv')
df
```

```python
plt.xlabel('Area(sqr ft)')
plt.ylabel('Price(US$)')
plt.scatter(df.area, df.price, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3300]]) 
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
d = pd.read_csv('Areas.csv')
d
```

```python
predicted_price = reg.predict(d)
d['Price'] = predicted_price
```

```python
d
```

```python
d.to_csv("Price.csv", index=False)
```

```python
plt.xlabel('Area(sqr ft)', fontsize=20)
plt.ylabel('Price(US$)', fontsize=20)
plt.scatter(df.area, df.price, color='red')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```
