```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('Canada Per Capita Income.csv')
df
```

```python
plt.xlabel('Year', fontsize=20)
plt.ylabel('Per Capita Income (US$)', fontsize=20)
plt.title('Canada Per Capita Income', fontsize=20)
plt.grid()
plt.plot(df.year, df['per capita income (US$)'], color='red')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])
```

```python
reg.predict([[2025]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
pred = pd.read_csv('pred_year.csv')
pred
```

```python
predicted_income = reg.predict(pred)
pred['Income'] = predicted_income
```

```python
pred
```

```python
pred.to_csv('predicted_income.csv', index=False)
```

```python
plt.xlabel('Year', fontsize=20)
plt.ylabel('per capita income (US$)', fontsize=20)
plt.scatter(df['year'], df['per capita income (US$)'], color='red')
plt.plot(df['per capita income (US$)'], reg.predict(df[['per capita income (US$)']]))
```
