```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv(
    'income.csv',
    index_col=None,
    names=['income', 'count'],
    skiprows=1
    )
df
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
g = sns.barplot(x='income', y='count', data=df)
g.set_xticklabels(g.get_xticklabels(), rotation=90, horizontalalignment='left')
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
g = sns.barplot(x='income', y='count', data=df)
g.set_xticklabels(g.get_xticklabels(), rotation=90, horizontalalignment='left')
g.set(xscale='log')
```
